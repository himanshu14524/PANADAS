```python
Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.

ANS-
In Pandas, a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure 
with labeled axes (rows and columns). It is essentially a collection of Series where each column of a DataFrame
can be thought of as a Series. DataFrames can store data of different types and are very similar to a spreadsheet
or SQL table.

On the other hand, a Pandas Series is a one-dimensional labeled array capable of holding data of any type. It can 
be seen as a single column of a DataFrame. Each element in a Series is associated with a unique label called an index.
DATAFRAME EXAMPLE :
    import pandas as pd

# Creating a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 28],
    'Gender': ['Female', 'Male', 'Male']
}

df = pd.DataFrame(data)
print("DataFrame:")
print(df)


SERIES EXAMPLE :
    import pandas as pd

# Creating a Series
ages = pd.Series([25, 30, 28], name='Age')
print("Series:")
print(ages)
```

```python
Q5. What are some common functions you can use to manipulate data in a Pandas DataFrame? Can
you give an example of when you might use one of these functions?

ANS-1.  head() and tail(): These methods allow you to view the first or last few rows of your DataFrame, respectively.
They're handy for a quick look at the data.

EXAMPLE: import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emily'],
    'Age': [25, 30, 28, 32, 29],
    'Gender': ['Female', 'Male', 'Male', 'Male', 'Female']
}

df = pd.DataFrame(data)

print("First few rows:")
print(df.head())  # Display first 5 rows

2.describe(): Provides descriptive statistics about the numerical columns in the DataFrame like count, mean, min, max, etc.

EXAMPLE:
    print("Descriptive statistics:")
print(df.describe())

3. drop(): Removes rows or columns from the DataFrame based on specified labels.

Example (dropping a column):
df_without_age = df.drop('Age', axis=1)  # axis=1 for columns, axis=0 for rows
print("DataFrame without 'Age' column:")
print(df_without_age)

4. groupby(): Groups data based on specified columns, allowing the application of aggregate functions on these groups.

Example (grouping by 'Gender' and finding average age):
    grouped_by_gender = df.groupby('Gender')['Age'].mean()
print("Average age by gender:")
print(grouped_by_gender)

5. apply() and applymap(): apply() allows applying a function along an axis of the DataFrame, while applymap() applies a function element-wise to a DataFrame.

Example (apply() to calculate the length of names):
    df['Name_Length'] = df['Name'].apply(len)
print("DataFrame with name lengths:")
print(df)
```
